# Should I stay in U.S ? Let's anylisis the unemploy rate

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import datetime

unrate=pd.read_csv("UNRATE.csv")
unrate['DATE']=pd.to_datetime(unrate['DATE'])
plt.plot(unrate['DATE'],unrate['UNRATE'])
plt.xticks(rotation=90)
plt.show()
```

# 首先，请减少数据量，用当年的月平均值代替12个月的数据 
   提示，可以利用pandas里的loc，tolist 方法
   利用numpy里的mean()方法求每年12个月的平均值

## 先计算一年的，以1949为例

```python
print(unrate.head(24))
print("Just print 1949")
unrate1949=unrate.loc[12:23]
print(unrate1949)

```

## 将12月的失业率转换为向量
提示：利用numpy的array方法


```python
import numpy as ny

vector1949=ny.array(unrate1949['UNRATE'])
print(vector1949)
```

## 然后求12个月的平均值
提示：利用numpy的mean()函数

```python
mean1949=vector1949.mean()
print(mean1949)
```

# 好，现在请将1948年到2017年的年平均值写到列表中名称为ur70years

提示：列表存储70年的年代数字
1.range(0,70) 生成 0到70的序列
2.用for循环遍历
3.用ur70years列表存储70年的失业率
4.用uy70years列表存储70年的年代数字
5.用ny.array 将当年的12个月转化为向量，名称为vector12month
6.用mean()求平均
7.用append 加入ur70years,uy70years

```python
ur70years=[]
uy70years=[]

for idx in range(0,70):
    year=1948+idx
    unreate12month=unrate.loc[idx*12:idx*12+11]
    vector12month=ny.array(unreate12month['UNRATE'])
    rate12monthmean=vector12month.mean()
    uy70years.append(year)
    ur70years.append(rate12monthmean)
```

# 请用matplotlib画出 unrate70years的曲线图

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.plot(uy70years,ur70years)
plt.xticks(rotation=90)
plt.show()

#print(uy70years)
#print(ur70years)
```
